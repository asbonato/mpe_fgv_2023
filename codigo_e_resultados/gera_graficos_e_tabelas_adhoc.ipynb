```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import random
seed = 42
random.seed(seed)
np.random.seed(seed)

sns.set()
sns.set_context("notebook")
sns.set(rc={'figure.figsize':(10,7)})
pd.set_option('display.float_format', '{:,.2f}'.format)
```

```python
import warnings
warnings.filterwarnings('ignore')
warnings.simplefilter('ignore')
```

```python
df_arr = pd.read_excel("variaveis_previsao.xlsx")
df_arr.head()
```

```python
df_arr.tail()
```

```python
mes = ['Jan','Fev','Mar','Abr','Mai','Jun','Jul','Ago','Set','Out','Nov','Dez']
df_arr['date'].apply(lambda x: str(mes[x.month-1])+'/'+str(x.year))
```

```python
df_icms = df_arr[['date','icms_sp']]
df_icms.index = df_arr['date']
df_icms['date'] = df_icms['date'].apply(lambda x: str(mes[x.month-1])+'/'+str(x.year))
df_icms.rename(columns={'date':'Data', 'icms_sp':'Valor Nominal R$'}, inplace=True)
df_icms
```

```python
df_esq = pd.DataFrame()
df_dir = pd.DataFrame()
```

```python
df_esq = df_esq.append(
    df_icms['2003-01-01':'2004-12-01']).append(
    df_icms['2007-01-01':'2009-12-01']).append(
    df_icms['2013-01-01':'2015-12-01']).append(
    df_icms['2019-01-01':'2020-12-01'])
```

```python
df_dir = df_dir.append(
    df_icms['2005-01-01':'2006-12-01']).append(
    df_icms['2010-01-01':'2012-12-01']).append(
    df_icms['2016-01-01':'2018-12-01']).append(
    df_icms['2021-01-01':'2022-12-01'])
df_dir.index = df_esq.index
```

```python
df_saida = df_esq.copy()
df_saida['branco1'] = ' '
df_saida['branco2'] = ' '
df_saida['Data1'] = df_dir['Data']
df_saida['Valor Nominal R$1'] = df_dir['Valor Nominal R$']
df_saida
```

```python
texto = df_saida['2003-01-01':'2004-12-01'].to_latex(index=False)
texto = texto.replace('.','_').replace(',','.').replace('_', ',')
print(texto)
```
